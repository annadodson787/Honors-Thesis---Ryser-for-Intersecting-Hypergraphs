```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dumb_lambda(x):
    for i in range(len(x)):
        element = x[i]
        if (element != None):
            x[i] = int(element)
    return x
```

```python
df = pd.read_csv('gamescores.csv', header=None, sep='\n')
df = df[0].str.split(',"', expand=True)
df = df.apply(lambda x: x.replace('\"', '', regex = True))
df = df.apply(lambda x: dumb_lambda(x), axis=0)
#df[0] = df[0].str.split(',"', expand=True)

```

```python
df.head()
```

```python
def lent(x):
    count = 0
   # print(x)
    for a in range(len(x)-1):
        if x[a] != None:
            count = count +1
            
    return count

df['length'] = df.apply(lambda x: lent(x),axis=1)
```

```python
df.head()
```

```python
df_is = []
for i in range(6,14):
    df_is.append(df[df['length'] == i])

print(df_is)
```

```python
sm = 0
for df_i in df_is:
    print(df_i.shape[0])
    sm = sm + df_i.shape[0]
    df_i=df_i.drop(['length'],axis=1)
    for row in range(len(df_i[0])):
        
        #print(df_i.iloc[row])
        plt.plot(df_i.iloc[row])
    plt.figure()

print(sm)
```

```python
def bool_lambda(x):
    for i in range(len(x)):
        element = x[i]
        if (element != None):
            if element == "true":
                x[i] = 1
            else:
                x[i] = 0
    return x

df_greedy = pd.read_csv('greedy.csv', header=None, sep='\n')
df_greedy = df_greedy[0].str.split(',"', expand=True)
df_greedy = df_greedy.apply(lambda x: x.replace('\"', '', regex = True))
df_greedy = df_greedy.apply(lambda x: bool_lambda(x), axis=0)
```

```python
df_greedy['length'] = df_greedy.apply(lambda x: lent(x),axis=1)
df_greedy.head(10)
```

```python
df_greedy_is = []
for i in range(7,14):
    df_greedy_is.append(df_greedy[df_greedy['length'] == i])

print(df_greedy_is)
```

```python
for df_greedy_i in df_greedy_is:
    print(df_greedy_i.shape[0])
    sm = sm + df_greedy_i.shape[0]
    df_greedy_i=df_greedy_i.drop(['length'],axis=1)
    for row in range(len(df_greedy_i[0])):
        
        #print(df_i.iloc[row])
        plt.plot(df_greedy_i.iloc[row].cumsum())
    plt.figure()
```
